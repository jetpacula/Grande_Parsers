```python
import csv 
with open('./labels.csv', 'r') as f:
    csvreader = csv.reader(f)
    labels = [i for i in csvreader]
    # using csv.writer method from CSV package
   # write = csv.writer(f)
   # write.readrows(labelling)
```

```python
from youtube_transcript_api import YouTubeTranscriptApi
  
# assigning srt variable with the list
# of dictonaries obtained by the get_transcript() function
srt = YouTubeTranscriptApi.get_transcript("hJzNWVkW83I")
  
# prints the result
for idx,obj in enumerate(srt):
    srt[idx]['text'] = obj['text'].replace("\n"," ")
    
    print(obj)
#print(srt)


 
```

```python
caps = []
for i in srt:
    caps.append([range (int(i['start']*1000),int(i['start']*1000 + i['duration']*1000))])
  #  print(i['start'],i['start'] + i['duration'])
 #   i['speaker'] = 


```

```python
speakers =[]
for i in labels:
    speakers.append([i[0],range(int(float(i[1])*1000),int(float(i[2])*1000))])
```

```python
for x in caps:
    zeros = 0
    ones = 0
    compareDict = {'0':0,'1':0}
    for y in speakers:
        xs = set(x[0])
        res = xs.intersection(y[1]) 
        if len(res)>0:
            compareDict[y[0]] += len(res)
        
            x.append([y[0],y[1],res])
    x.append()    
            
```
